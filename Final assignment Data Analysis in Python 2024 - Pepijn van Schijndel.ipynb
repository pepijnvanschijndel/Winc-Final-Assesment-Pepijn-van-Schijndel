### Step 1 Define data
To define which country makes the biggest strides in decreasing CO2 output, I have to calculate the relative decrease in CO2 emissions over time for each country and identify the country with the highest percentage decrease. As I already use the CO2 emission per capita data, it will be relative to the population as well.

```python
import pandas as pd

# Use the df_pred DataFrame from the first assigment
# Calculate the average CO2 emission for each country
df_avg_country = df_pred.groupby(['Country', 'Year']).mean().reset_index()
```

### Step 2 Combine data
Use a pivot table to get years as columns and countries as rows from the dataset from Final Assigment - 1.

```python
# Pivot the DataFrame to get years as columns and countries as rows
df_pivot = df_avg_country.pivot(index='Country', columns='Year', values='CO2 emission')

# Calculate the percentage change from the first to the last year for each country
df_pivot['Percentage Change'] = (df_pivot[df_pivot.columns[-1]] - df_pivot[df_pivot.columns[0]]) / df_pivot[df_pivot.columns[0]] * 100

# Sort countries by the percentage change
df_sorted = df_pivot.sort_values(by='Percentage Change')
```

```python
# Print the country with the biggest decrease in CO2 emissions
country_with_biggest_decrease = df_sorted.index[0]
percentage_decrease = df_sorted.iloc[0]['Percentage Change']

print(f"The country with the biggest decrease in CO2 emissions is {country_with_biggest_decrease} with a decrease of {percentage_decrease:.2f}%.")
```

### Step 3 Make a graph

```python
# Remove entities with no value in Percentage Change
df_pivot = df_pivot.dropna(subset=['Percentage Change'])

# Sort countries by the percentage change in descending order
df_sorted = df_pivot.sort_values(by='Percentage Change', ascending=False)

# Plot the data
plt.figure(figsize=(14, 8))
plt.barh(df_sorted.index, df_sorted['Percentage Change'], color='skyblue')

# Reverse the x-axis
plt.gca().invert_xaxis()

plt.xlabel('Percentage Change in CO2 Emissions')
plt.ylabel('Country')
plt.title('Countries with Highest Decrease in CO2 Emissions Percentage')
plt.grid(axis='x')

# Show the plot
plt.show()
```

#### Step 4 Conclusion
The country with the biggest decrease in CO2 emissions is Denmark with a decrease of -39.19%.

## Step 3 Data Preprocessing

Ingest, clean and preprocess the data, handling missing values, outliers, and normalizing the features if necessary.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Load the dataset
file_path = '/content/drive/MyDrive/Colab Notebooks/Winc Final Assigment - Data analysis in Python - Pepijn van Schijndel/Datasets/costs-of-66-different-technologies-over-time.csv'
df = pd.read_csv(file_path)

# Filter the dataframe for the specific entities
entities = ['Concentrating solar', 'Geothermal electricity', 'Photovoltaics', 'Wind turbine (Denmark)']
df_filtered = df[df['Entity'].isin(entities)]

# Further filter the data to include only years from 1981 onwards
df_filtered = df_filtered[df_filtered['Year'] >= 1981]

# Plot the data and linear regression lines
plt.figure(figsize=(14, 8))

for entity in entities:
    entity_df = df_filtered[df_filtered['Entity'] == entity]
    x = entity_df['Year']
    y = entity_df['Technology cost']

    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)

    # Plot the data points
    plt.scatter(x, y, label=f'{entity} data points')

    # Plot the regression line and extend it to 2026
    x_pred = np.arange(1981, 2027)
    y_pred = slope * x_pred + intercept
    plt.plot(x_pred, y_pred, label=f'{entity} regression line')

# Add a dotted black zero line
plt.axhline(y=1, color='black', linestyle='--', label='Zero Cost Line')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Technology cost')
plt.yscale('log')  # Use logarithmic scale for better visualization
plt.title('Linear Regression Lines for Different Renewable Energy Technologies (1981-2026)')
plt.legend()

# Show the plot
plt.show()
```
